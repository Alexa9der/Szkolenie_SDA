```python
import seaborn as sns
penguins = sns.load_dataset('penguins')

penguins.head()
```

1. wyswietl 10 obserwacji species, island i bill_length_mm z tabeli penguins


```python
query = """SELECT species, island , bill_length_mm
            FROM penguins  LIMIT 10"""
pysqldf(query)
```

2. wyswietl te samice pangiwna dla ktorych masa ciala wynosi wiecej niz 4000


```python
query = """SELECT species, island , bill_length_mm
            FROM penguins  
            where body_mass_g > 4000 AND sex == "Female"
            LIMIT 10"""
pysqldf(query)
```

3. wyswietl plec, mase i gatunek pingwina dla ktorych bill_depth_mm wynosi mniej niz 15. 
 Posortuj wyniki po masie zwierzaka

```python
query = """SELECT sex, body_mass_g , species
            FROM penguins  
            where bill_depth_mm < 15
            ORDER BY body_mass_g """
pysqldf(query)
```

4. Wyswietl maksymalna wartosc wagi w zaleznosci od plci i gatunku pingwina


```python
query = """SELECT sex, species,  MAX(body_mass_g) AS max_body_mass_g
            FROM penguins
            GROUP BY sex, species """
pysqldf(query)
```

5. Oblicz stosunek dlugosci (bill_length_mm)  do glebokosci dzioba (bill_depth_mm)

```python
query = """
            SELECT bill_length_mm / bill_depth_mm AS result
            FROM penguins ;
"""
pysqldf(query)
```
