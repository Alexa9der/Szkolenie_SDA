# Zad 9. Stw√≥rz funkcjƒô, kt√≥ra przyjmuje dwa wektory (oba o tym samym wymiarze) w postaci numpy tablicy, a zwr√≥ci odleg≈Ço≈õƒá euklidesowƒÖ miƒôdzy nimi.


```python
v = np.random.randint(1,10,10).reshape(5,2)
v2 =  np.random.randint(1,10,10).reshape(5,2)

def euclidean_distance (vector: np.array, vector_2: np.array) -> np.array:
    if len(vector) == len(vector_2):
        distance = [ np.sqrt(sum( (vector[i] - vector_2[i])**2)) for i in range(len(vector)) ]
        return distance
    else: 
        return "Length vector does not match"

print(euclidean_distance(v,v2))
```

```python
m = 175
sigma = 12

df = pd.DataFrame({"wzrost": np.random.normal(loc = m, scale = sigma, size = 20)})
df.describe()
```

* m0 = 175
* H0: m = m0

```python
m0 = 175
T, pvalue = sts.ttest_1samp(df.wzrost, popmean = m0, alternative = "two-sided")
print("T:", T, "p-value:", pvalue)
```

```python
n_df = n - 1
(1 - sts.t.cdf(x = np.abs(T_calc), df = n_df)) * 2
```

###### Test t-studenta do weryfikacji hipotez o ≈õredniej dla dw√≥ch populacji
###### UWAGA ! W przypadku du≈ºych pr√≥b (najczƒô≈õciej powy≈ºej 30) rozk≈Çad poni≈ºszej statystyki testowej zbiega do rozk≈Çadu normalnego. W zwiƒÖzku z tym mo≈ºna aproksymowaƒá (przybli≈ºaƒá) rozk≈Çad t-studenta rozk≈Çadem normalnym


```python
m1, m2 = 175, 180
sigma1, sigma2 = 10, 10
n1, n2 = 20, 20
prob1 = pd.DataFrame({"wzrost": np.random.normal(loc = m1, scale = sigma1, size = n1)})
prob2 = pd.DataFrame({"wzrost": np.random.normal(loc = m2, scale = sigma2, size = n2)})

print(prob1.mean()[0], prob2.mean()[0])
```

* H0: m1 = m2
* H1: m1 != m2

```python
T, pvalue = sts.ttest_ind(a = prob1.wzrost, b = prob2.wzrost, alternative = "two-sided")
print("T:", T, "p-value:", pvalue)
```

```python
xsr1, xsr2 = np.mean(prob1.wzrost), np.mean(prob2.wzrost)
var1, var2 = np.var(prob1.wzrost), np.var(prob2.wzrost)

T_calc = (xsr1 - xsr2) / np.sqrt((n1 * var1 + n2 * var2)/(n1 + n2 - 2)*(1 / n1 + 1 / n2))
T_calc
```

```python
n_df = len(prob1.wzrost) + len(prob2.wzrost) - 2
(1 - sts.t.cdf(x = np.abs(T_calc), df = n_df)) * 2
```

* H0: m1 = m2
* H1: m1 < m2

```python
T, pvalue = sts.ttest_ind(a = prob1.wzrost, b = prob2.wzrost, alternative = "less")
print("T:", T, "p-value:", pvalue)
```

```python
sts.t.cdf(x = T_calc, df = n_df)
```

* H0: m1 = m2
* H1: m1 > m2

```python
T, pvalue = sts.ttest_ind(a = prob1.wzrost, b = prob2.wzrost, alternative = "greater")
print("T:", T, "p-value:", pvalue)
```

```python
1 - sts.t.cdf(x = T_calc, df = n_df)
```

###### Testy permutacyjne jako alternatywa dla klasycznych test√≥w parametrycznych - weryfikacja hipotez o ≈õrednich w dw√≥ch populacjach
###### UWAGA ! Testy permutacyjne nie wymagajƒÖ spe≈Çnienia za≈Ço≈ºe≈Ñ kt√≥re towarzyszƒÖ klasycznym testom

```python
m1, m2 = 175, 180
sigma1, sigma2 = 10, 10
n1, n2 = 20, 20
prob1 = pd.DataFrame({"wzrost": np.random.normal(loc = m1, scale = sigma1, size = n1)})
prob2 = pd.DataFrame({"wzrost": np.random.normal(loc = m2, scale = sigma2, size = n2)})

print(prob1.mean()[0], prob2.mean()[0])
```

```python
def perm_test_m(var1, var2, nperm = 300):
    n1 = len(var1)
    n2 = len(var2)
    n = n1 + n2
    T0 = np.abs(np.mean(var1) - np.mean(var2))
    T1 = []
    vrs = pd.concat([var1, var2]).values
    
    for i in range(nperm):
        vrs = np.random.choice(vrs, size = n, replace = False)
        T1.append(np.abs(np.mean(vrs[:n1]) - np.mean(vrs[n1:])))
        
    print("P-value:", np.round(np.mean(pd.Series(T1) >= T0), 5))
    sns.histplot(T1)
    plt.axvline(T0, color = "red")
    plt.show()
```

```python
perm_test_m(prob1.wzrost, prob2.wzrost, nperm = 1000)
```

```python
x = np.random.normal(loc = 175, scale = 10, size = 30)
y = np.random.normal(loc = 175, scale = 20, size = 30)
z = np.random.normal(loc = 175, scale = 30, size = 30)

T, pvalue = sts.levene(x, y, z, center = "mean")

print("T:", T, "p-value:", pvalue)
```

```python
ex = np.abs(x - np.mean(x))
ey = np.abs(y - np.mean(y))
ez = np.abs(z - np.mean(z))

k = 3
n1 = len(ex)
n2 = len(ey)
n3 = len(ez)
n = n1 + n2 + n3
exsr = np.mean(ex)
eysr = np.mean(ey)
ezsr = np.mean(ez)
esr = (exsr * n1 + eysr * n2 + ezsr * n3) / n


nom = n1 * np.power(exsr-esr, 2) + n2 * np.power(eysr-esr, 2) + n3 * np.power(ezsr-esr, 2)
denom = np.sum(np.power(ex - exsr, 2)) + np.sum(np.power(ey - eysr, 2)) + np.sum(np.power(ez - ezsr, 2))

T_calc = ((n - k) / (k - 1)) * (nom / denom)
T_calc
```

```python
1 - sts.f.cdf(x = T_calc, dfn = k - 1, dfd = n - k)
```

```python
k = 3
nx = len(x)
ny = len(y)
nz = len(z)
n = nx + ny + nz
varx = np.var(x)
vary = np.var(y)
varz = np.var(z)
varp = ((nx - 1) * varx + (ny - 1) * vary + (nz - 1) * varz) / (n - k)

nom = (n - k) * np.log(varp) - ((nx - 1) * np.log(varx) + (ny - 1) * np.log(vary) + (nz - 1) * np.log(varz))
denom = 1+((1/(3*(k-1)))*(1/(nx-1)+1/(ny-1)+1/(nz-1)) - 1/(n-k))

T_calc = nom / denom
T_calc
```

###### Testy permutacyjne jako alternatywa dla klasycznych test√≥w r√≥wno≈õci wariancji

```python
x = pd.Series(np.random.normal(loc = 175, scale = 10, size = 30))
y = pd.Series(np.random.normal(loc = 175, scale = 15, size = 30))
```

```python
def perm_test_s(var1, var2, nperm = 300):
    n1 = len(var1)
    n2 = len(var2)
    n = n1 + n2
    T0 = np.abs(np.std(var1) - np.std(var2))
    T1 = []
    vrs = pd.concat([var1, var2]).values
    
    for i in range(nperm):
        vrs = np.random.choice(vrs, size = n, replace = False)
        T1.append(np.abs(np.std(vrs[:n1]) - np.std(vrs[n1:])))
        
    print("P-value:", np.round(np.mean(pd.Series(T1) >= T0), 5))
    sns.histplot(T1)
    plt.axvline(T0, color = "red")
    plt.show()
```

```python
perm_test_s(x, y, nperm = 1000)
```
